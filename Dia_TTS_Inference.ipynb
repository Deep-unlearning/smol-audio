# Dialogue TTS with Dia-1.6B

This notebook demonstrates how to generate realistic multi-speaker dialogue using the Dia-1.6B model from nari-labs. Dia is a text-to-speech model specifically designed for generating natural-sounding conversations with multiple speakers, making it ideal for creating dialogue audio, podcasts, or interactive voice applications.

> **Requirements**: Dia requires **Python 3.10+** and `numpy>=2.2.4`. Run this notebook on Google Colab or a compatible environment.

## Load Model

```python
import torch
import soundfile as sf
from IPython.display import Audio
import numpy as np
from dia.model import Dia

# Load the Dia-1.6B model
print("Loading Dia-1.6B...")
model = Dia.from_pretrained("nari-labs/Dia-1.6B", compute_dtype="float16")

print("Model loaded successfully!")
```

## Generate Dialogue Speech

```python
# Create a dialogue with speaker tags
dialogue_text = """[S1] Hello! How are you doing today?
[S2] I'm doing great, thanks for asking! How about you?
[S1] I'm wonderful. I wanted to talk to you about our upcoming project.
[S2] Sure, I'd love to hear more about it. What do you have in mind?"""

print("Generating dialogue audio...")
print(f"Input text:\n{dialogue_text}\n")

# Generate audio
audio_output = model.generate(dialogue_text, use_torch_compile=False, verbose=True)

print("Audio generated successfully!")
```

## Play Audio

```python
# Save the generated audio to a file
output_file = "dialogue_output.wav"
sample_rate = 44100  # Dia's output sample rate

# Save audio
sf.write(output_file, audio_output, sample_rate)
print(f"Audio saved to {output_file}")

# Play audio in notebook
Audio(audio_output, rate=sample_rate)
```

## Experiment with Different Dialogues

```python
# Try a different dialogue scenario
dialogue_text_2 = """[S1] Welcome to our podcast! Today we're discussing artificial intelligence.
[S2] Thanks for having me! I'm excited to be here.
[S1] Let's start with the basics. What exactly is machine learning?
[S2] Machine learning is a subset of AI where computers learn from data without being explicitly programmed.
[S1] That's fascinating! Can you give us a real-world example?
[S2] Sure! Think about recommendation systems on streaming platforms that learn your preferences over time."""

print("Generating second dialogue...")
audio_output_2 = model.generate(dialogue_text_2, use_torch_compile=False, verbose=True)

# Save and play
output_file_2 = "dialogue_output_2.wav"
sf.write(output_file_2, audio_output_2, sample_rate)
print(f"Audio saved to {output_file_2}")

Audio(audio_output_2, rate=sample_rate)
```

## Tips

### Speaker Tags Format
- Use `[S1]` and `[S2]` tags to indicate different speakers in your dialogue
- You can use additional speaker tags like `[S3]`, `[S4]` if the model supports them
- Place speaker tags at the beginning of each line or utterance
- Each speaker will have a distinct voice characteristic

### Generation Parameters
- **temperature**: Controls randomness (0.7-0.9 recommended for natural speech)
- **max_length**: Maximum length of generated audio (adjust based on dialogue length)
- **do_sample**: Set to True for more natural and varied output
- **top_p**: Can be used for nucleus sampling (try values like 0.9-0.95)

### Best Practices
- Keep individual utterances natural and conversational
- Avoid overly long monologues for better quality
- Use proper punctuation to guide prosody and pauses
- Experiment with different temperature values to find the right balance between consistency and naturalness
